# DF TRIPS 

## import libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

## import csv

```python
df_trips_url = "trips_df.csv"
```

```python
df_trips = pd.read_csv(df_trips_url)
```

```python
df_trips
```

## info about df

```python
# Display the dimensions of the dataset
print("Dimensions of the dataset:", df_trips.shape)

# Check the data types of each column
print("\nData types of each column:")
print(df_trips.dtypes)

# Check for missing values
print("\nMissing values:")
print(df_trips.isnull().sum())

# Check for duplicates
print("\nNumber of duplicate rows:", df_trips.duplicated().sum())
```

## clean the df

```python
# Impute missing numerical values with mean
numeric_cols = df_trips.select_dtypes(include=['int', 'float']).columns
df_trips[numeric_cols] = df_trips[numeric_cols].fillna(df_trips[numeric_cols].mean())

# Impute missing categorical values with mode
categorical_cols = df_trips.select_dtypes(include=['object']).columns
df_trips[categorical_cols] = df_trips[categorical_cols].fillna(df_trips[categorical_cols].mode().iloc[0])

# Verify if missing values have been handled
print("\nMissing values after imputation:")
print(df_trips.isnull().sum())
```

```python
# Drop rows with missing values in the "Start Date Time(s)" column
df_trips.dropna(subset=['Start Date Time(s)'], inplace=True)

# Verify if missing values have been dropped
print("\nMissing values after dropping rows:")
print(df_trips.isnull().sum())
```

```python
df_trips.info()
```

## convert date

```python
# Convert Start Date Time and End Date Time to datetime objects
df_trips['start_date_time'] = pd.to_datetime(df_trips['Start Date Time(s)'])
```

```python
df_trips.info()
```

```python
df_trips['start_date_time'] = df_trips['start_date_time'].astype(str)
df_trips['date'] = df_trips['start_date_time'].str.split(' ').str[0]
df_trips['time'] = df_trips['start_date_time'].str.split(' ').str[1]

# Verify the new columns
print("\nNew columns added:")
print(df_trips[['date', 'time']].head())
```

```python
print(df_trips[['date', 'time']].tail())
```

```python
df_trips.drop(columns=['Start Date Time(s)', ' Customer Last Name(s)', ' Customer First Name(s)'], inplace=True)

# Verify that the columns have been deleted
print("\nColumns deleted:")
print(df_trips.head())
```

```python
# vehicle number and vehicle type separately
df_trips[['vehicle_number', 'vehicle_type']] = df_trips[' Scheduled Vehicle'].str.extract(r'(\d+)-?\s*(\w+)')
print(df_trips[['vehicle_number', 'vehicle_type']].head())
```

```python
print(df_trips[['vehicle_number', 'vehicle_type']].tail())
```

```python
df_trips.tail()
```

# Vehicles

### mapping vehicle type

```python
df_trips['vehicle_type'].unique()
```

```python
df_trips[' Scheduled Vehicle'].unique()
```

```python
# Count the number of rows where 'vehicle_type' is 'Winnipeg WAV'
count_wav = df_trips[df_trips[' Scheduled Vehicle'] == ' Winnipeg WAV'].shape[0]

print("Number of lines with 'Winnipeg WAV' in 'vehicle_type':", count_wav)
```

```python
# Remove leading and trailing white spaces in the 'Scheduled Vehicle' column
df_trips[' Scheduled Vehicle'] = df_trips[' Scheduled Vehicle'].str.strip()

# Verify the column after removing white spaces
print("\nScheduled Vehicle column after removing white spaces:")
print(df_trips[' Scheduled Vehicle'].head())
```

```python
print(df_trips[' Scheduled Vehicle'].unique())
```

```python
df_trips[' Scheduled Vehicle']
```

```python
def determine_vehicle_type(vehicle):
    if 'BUS' in vehicle:
        return 'BUS'
    elif 'CAR' in vehicle:
        return 'CAR'
    elif 'VAN' in vehicle:
        return 'VAN'
    elif 'Winnipeg WAV' in vehicle:
        return 'Winnipeg WAV'
    else:
        return None  

# Apply function to create 'vehicle_type' column
df_trips['vehicle_type'] = df_trips[' Scheduled Vehicle'].str.strip().apply(determine_vehicle_type)

# Verify the new 'vehicle_type' column
print(df_trips[[' Scheduled Vehicle', 'vehicle_type']].tail(100))

```

```python
df_trips[[' Scheduled Vehicle', 'vehicle_type']].tail(100)
```

```python
# Define mapping dictionary for vehicle types
type_mapping = {
    'BUS': 1,
    'CAR': 2,
    'VAN': 3,
    'Winnipeg WAV': 4
}

# Create a new column 'vehicle_type_mapped' by mapping the extracted vehicle types
df_trips['vehicle_type_mapped'] = df_trips['vehicle_type'].map(type_mapping)

# Verify the new columns
print("\nNew columns 'vehicle_type' and 'vehicle_type_mapped' added:")
print(df_trips[[' Scheduled Vehicle', 'vehicle_type', 'vehicle_type_mapped']].tail(100))
```

### plot Distribution of the vehicle tyoes

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df_trips, x='vehicle_type', order=df_trips['vehicle_type'].value_counts().index)
plt.title('Distribution of Vehicle Types 2023/2024')
plt.xlabel('Vehicle Type')
plt.ylabel('Count')
plt.xticks(rotation=45)

for index, value in enumerate(df_trips['vehicle_type'].value_counts()):
    plt.text(index, value + 5, str(value), ha='center', va='bottom')

plt.show()
```

```python
# Extract hour from the start date time
df_trips['hour'] = pd.to_datetime(df_trips['start_date_time']).dt.hour
```

```python
hourly_vehicle_counts = df_trips.groupby(['hour', 'vehicle_type']).size().reset_index(name='trip_count')

plt.figure(figsize=(12, 6))
sns.lineplot(data=hourly_vehicle_counts, x='hour', y='trip_count', hue='vehicle_type', marker='o')
plt.title('Count of Trips by Hour for Each Vehicle Type')
plt.xlabel('Hour of Day')
plt.ylabel('Count of Trips')
plt.xticks(range(24))
plt.legend(title='Vehicle Type')
plt.show()
```

```python

week_vehicle_counts = df_trips.groupby(['weekday', 'vehicle_type']).size().reset_index(name='trip_count')

plt.figure(figsize=(12, 6))
sns.lineplot(data=week_vehicle_counts, x='weekday', y='trip_count', hue='vehicle_type', marker='o')
plt.title('Count of Trips by weekday for Each Vehicle Type')
plt.xlabel('Weekday')
plt.ylabel('Count of Trips')
plt.xticks(range(7), ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.legend(title='Vehicle Type')
plt.show()
```

```python
monthly_vehicle_counts = df_trips.groupby(['month', 'vehicle_type']).size().reset_index(name='trip_count')

plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_vehicle_counts, x='month', y='trip_count', hue='vehicle_type', marker='o')
plt.title('Count of Trips by Month for Each Vehicle Type')
plt.xlabel('Month')
plt.ylabel('Count of Trips')
plt.xticks(range(1, 13), ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'])
plt.legend(title='Vehicle Type')
plt.show()
```

```python
pickup_trips = df_trips[['vehicle_type', ' Origin Address(s)']]
dropoff_trips = df_trips[['vehicle_type', ' Destination Address(s)']]

# Group the data by vehicle type and address for both pick-up and drop-off trips
pickup_counts = pickup_trips.groupby(['vehicle_type', ' Origin Address(s)']).size().reset_index(name='pickup_count')
dropoff_counts = dropoff_trips.groupby(['vehicle_type', ' Destination Address(s)']).size().reset_index(name='dropoff_count')

# Find the top 20 addresses for each vehicle type for both pick-up and drop-off
top_pickup_addresses = pickup_counts.groupby('vehicle_type').apply(lambda x: x.nlargest(20, 'pickup_count'))
top_dropoff_addresses = dropoff_counts.groupby('vehicle_type').apply(lambda x: x.nlargest(20, 'dropoff_count'))

# Display the top 20 addresses for each vehicle type for pick-up and drop-off
print("Top 20 Addresses for Pick-up:")
print(top_pickup_addresses)
print("\nTop 20 Addresses for Drop-off:")
print(top_dropoff_addresses)
```

```python
# Save the top 20 addresses for pick-up and drop-off to CSV files
top_pickup_addresses.to_csv('top_pickup_addresses.csv', index=False)
top_dropoff_addresses.to_csv('top_dropoff_addresses.csv', index=False)
```
